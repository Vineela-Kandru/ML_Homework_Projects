**Question 5**

* Let's find the least useful feature using the feature elimination technique.
* Train a model with all these features (using the same parameters as in Q4).
* Now exclude each feature from this set and train a model without it. Record the accuracy for each model.
* For each feature, calculate the difference between the original accuracy and the accuracy without the feature.

**Which of following feature has the smallest difference?**

***ANSWER - city_mpg***
 ***- Accuracy : 0.9513150531617236***
 ***- Difference : 0.0***

```python
from sklearn.feature_extraction import DictVectorizer

features = categories + numerical

for feature_to_eliminate in features:

    # Create a list of features excluding the current one
    Features_used = [feature for feature in features if feature != feature_to_eliminate]

# Creating Dictionaries of training dataset.
    train_dicts = df_train[Features_used].to_dict(orient = 'records')

#Initializing DictVectorizer
    dv = DictVectorizer(sparse = False)

# Training the DictVectorizer uing dictionary and Transforming
    X_train = dv.fit_transform(train_dicts)

# FOR VALIDATION DATASET

    val_dicts = df_val[categories + numerical].to_dict(orient = 'records')

    X_val = dv.transform(val_dicts)

# Logistic Regression

    from sklearn.linear_model import LogisticRegression

# initializing the model
    model = LogisticRegression(solver='liblinear', C=10, max_iter=1000, random_state=42)

#Training the model using training dataset
    model.fit(X_train, y_train)

# CALCULATING ACCURACY ON VALIDATION DATASET

    y_pred = model.predict(X_val)

    Accuracy = (y_val == y_pred).mean()

    print('without feature ' '"',feature_to_eliminate,'"')

    print('Accuracy :', Accuracy )

    print('Difference :' ,((Original_Accuracy - Accuracy)))

    print('\n')
```

**Question 6**

* **For this question, we'll see how to use a linear regression model from Scikit-Learn**.
* **We'll need to use the original column price. Apply the logarithmic transformation to this column.**
* **Fit the Ridge regression model on the training data with a solver 'sag'. Set the seed to 42.**
* **This model also has a parameter alpha. Let's try the following values: [0, 0.01, 0.1, 1, 10].**
* **Round your RMSE scores to 3 decimal digits.**

**Which of these alphas leads to the best RMSE on the validation set?**

***ANSWER - Alpha = 0***

```python
# Logarithmic transformation of price column

df.price = np.log1p(df['price'])
```

```python
df.price
```

```python
# Splitting and Training the model

#Splitting
df_full_train, df_test = train_test_split(df, test_size=0.2, random_state=42)
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=42)

#dropping indices
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)

#Target variables
y_train = df_train.price.values
y_val = df_val.price.values
y_test = df_test.price.values

#Deleting target variable from the three datasets
del df_train['price']
del df_val['price']
del df_test['price']
```

```python
#Fitting and Transforming using DictVectorizer

train_dict = df_train[categories + numerical].to_dict(orient='records')

dv = DictVectorizer(sparse=False)
dv.fit(train_dict)

X_train = dv.transform(train_dict)

val_dict = df_val[categories + numerical].to_dict(orient='records')
X_val = dv.transform(val_dict)
```

```python
#Fitting the Ridge Regression model on the training data

from sklearn.linear_model import Ridge

# Calculating the RMSE

def rmse(y,y_pred):
  error = y - y_pred
  se = error**2
  mse = se.mean()
  return np.sqrt(mse)


for alpha in [0, 0.01, 0.1, 1, 10]:
    model = Ridge(alpha=alpha, solver="sag", random_state=42)
    model.fit(X_train, y_train)

    y_pred = model.predict(X_val)

    score = rmse(y_val, y_pred)

    print(alpha,'- rmse : ', score.round(3))

```
